```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Example data: time in days and tumor volume in mL
time = np.array([0, 30, 60, 90, 120])
volume = np.array([10, 15, 25, 40, 65])

# Exponential growth model
def exp_growth(t, V0, r):
    return V0 * np.exp(r * t)

# Fit the model
popt, _ = curve_fit(exp_growth, time, volume, p0=[volume[0], 0.01])
V0, r = popt
print(f"Initial Volume V0 = {V0:.2f} mL, Growth Rate r = {r:.4f} per day")
print(f"Doubling Time â‰ˆ {np.log(2)/r:.1f} days")

# Predict future growth
future_time = np.linspace(0, 180, 100)
predicted_volume = exp_growth(future_time, *popt)

# Plot
plt.scatter(time, volume, label="Observed", color="red")
plt.plot(future_time, predicted_volume, label="Predicted Growth")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mL)")
plt.title("Brain Tumor Growth Prediction")
plt.legend()
plt.show()
```
